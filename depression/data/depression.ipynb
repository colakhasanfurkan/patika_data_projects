```python
#intialize libraires and read data
import pandas as pd
import numpy as np
import math
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("/media/hasanfurkan/Data/PY/patika_data_projects/depression/data/scores.csv")

df.head()
```

```python
#split data
df = df.iloc[:23,1:12]
df
```

```python
#clean missing data
df.isnull().sum()
for i in df["melanch"].values:
    df['melanch'] = np.where(df["melanch"].isnull(), df["melanch"].mean(), df['melanch'])
df['edu'] = np.where(df['edu'] == " ", 3, df['edu'])
```

```python
#categorize data
df['edu'] = np.where(df['edu'] == "6-10", 1, df['edu'])
df['edu'] = np.where(df['edu'] == "11-15", 2, df['edu'])
df['edu'] = np.where(df['edu'] == "16-20", 3, df['edu'])

df['age'] = np.where(df['age'] == "45-49", 1, df['age'])
df['age'] = np.where(df['age'] == "50-54", 2, df['age'])
df['age'] = np.where(df['age'] == "35-39", 3, df['age'])
df['age'] = np.where(df['age'] == "25-29", 4, df['age'])
df['age'] = np.where(df['age'] == "30-34", 5, df['age'])
df['age'] = np.where(df['age'] == "20-24", 6, df['age'])
df['age'] = np.where(df['age'] == "40-44", 7, df['age'])
df['age'] = np.where(df['age'] == "60-64", 8, df['age'])
df['age'] = np.where(df['age'] == "65-69", 9, df['age'])
df['age'] = np.where(df['age'] == "55-59", 10, df['age'])
df
```

```python
#convert data types
df["age"] = df["age"].astype(int)
df["edu"] = df["edu"].astype(int)
df.dtypes
```

```python
#split data
x = df.iloc[:,2:12]
y = df.iloc[:,1:2]
```

```python
#split data for train and test
from fast_ml.model_development import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
#random forrest regression
regressor = RandomForestRegressor(n_estimators = 10, n_jobs=-1)
regressor.fit(x_train,y_train)
regressor.score(x_train,y_train)

def rmse(x,y):
    return math.sqrt(((x-2)**2).mean())

def print_score(m):
    print(f"RMSE of train set: {rmse(m.predict(x_train),y_train)}")
    print(f"RMSE of valid set: {rmse(m.predict(x_test),y_test)}")
    print(f"R^2 of train set: {m.score(x_train,y_train)}")
    print(f"R^2 of valid set: {m.score(x_test,y_test)}")
print(regressor.score(x_train,y_train))
print_score(regressor)
```
